```python
omdb = 'http://www.omdbapi.com/?apikey=7b298284&t='
```

```python
#add .head after weekend_box to limit API calls default number is 5
for index, row in weekend_box.iterrows():
    t = str(row['Title'])
    t = t.replace(":","")
    t = t.replace("'","")
    t = t.replace(' ', '+')
    search = omdb + t
    print (search)
    try:
        results = req.get(search).json()
        if results['Response'] == 'False':
            print("Couldn't find title: ", t)
        else:
            weekend_box.set_value(index, 'Director', results['Director'])
            weekend_box.set_value(index, 'imdbID', results['imdbID'])
            weekend_box.set_value(index, 'Actor', results['Actors'])
            weekend_box.set_value(index, 'Genre', results['Genre'])
            weekend_box.set_value(index, 'Writer', results['Writer'])
    except:
        print('in except')
        continue
```

```python
imdb = 'https://www.imdb.com/title/'
for index, row in weekend_box.head().iterrows():
    imdbID = row['imdbID']
    page = imdb + imdbID

    try:
        response = req.get(page)
        soup = BeautifulSoup(response.text, 'html.parser')
        txt_block = soup.find_all('div', class_="txt-block")
        budget_block = txt_block[9]
        budget = budget_block.text
        budget = budget.strip()
        budget = budget.split('\n')[0]
        budget = budget.split(':')[1]
        weekend_box.set_value(index, 'Budget', budget)
    except:
        continue
```
